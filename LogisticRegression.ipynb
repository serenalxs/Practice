```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/serenalxs/Practice/refs/heads/main/Basic/credit-a.csv'
data = pd.read_csv(url, header = None)
data.info()
```

sigmoid : output probability

      sigmoid = 1/ (1+e^(-x))

nn.SGDLoss() : MSE loss function (used in regression)

nn.BCELoss() : (binary cross entropy) Cross-Entropy Loss Function (used in logistic regression), enlarge the loss function making the optimization fast. The distance between actual distribution and calculated distribution.

      Loss = sum(p X log(q))

```python
print(data.head())
print ("--X---")
X = data.iloc[:,:-1]
print (X.head())
print ("---Y--")
Y = data.iloc[:,-1].replace(-1,0)
print (Y.head())

### convert to torch ###
print ("---tensor--")
X = torch.from_numpy(X.values).type(torch.float32)
Y = torch.from_numpy(Y.values.reshape(-1,1)).type(torch.float32)

print (X.shape)
print (Y.shape)
```

# **Build model**

nn.Sequential : add multiple layers

SGD optimizer : random gradient not consider prior gradient

Adam optimizer : sequential gradient considering prior gradients

```python
from torch import nn

model = nn.Sequential(
    nn.Linear (in_features = 15, out_features = 1) ,
    nn.Sigmoid ()        #activate sigmoid function to calculate probability

)

print ("-----model------")
print (model)

loss_fn = nn.BCELoss()
opt = torch.optim.Adam( model.parameters(), lr= 0.0001)
```

Batch -- split data to different batches

```python
batches = 16
no_of_batches = len(X) // batches
print ("-----calculate batches------")
print (no_of_batches)

epoches = 1000

for epoch in range(epoches) :
  for i in range(no_of_batches) :
    start = batches * i
    end = start + batches
    x = X [start : end]
    y = Y [start : end]
    y_pred = model(x)
    loss = loss_fn(y_pred,y)

    opt.zero_grad() #Clears the gradients from the previous step
    loss.backward() #Computes the gradients of the loss with respect to the parameters
    opt.step() #Updates the parameters using these gradients

print ("-----build model : sigmoid (y =b + w1x1+w2x2+...+w15x15------")
print (model.state_dict())
```

# **Calculate accuracy**

```python
( ( model(X).data.numpy() > 0.5 ).astype(int) == Y.numpy()).mean()
```

Activation function to enable non linear relationship:

RELU()

Sigmoid()

Tahn()

LeakRelu()

```python
#Relu: max (x,0)

x = torch.randn(10)
z = torch.relu(x)

print(x)
print("RELU activation:",z)

#sigmoid: limit to 0 ~ 1, used in logistical regression
z = torch.sigmoid(x)
print("sigmoid activation:", z)

#tanh: limit to -1 ~ 1
z = torch.tanh(x)
print("Tanh activation:", z)
```
