# **Gradient**


activate gradient : requires_grad = True

activate trace back : .backward() , only applied to scalar (number, not to vector)

look for **gradient** : .grad

look for **gradient function** : .grad_fn , the first derivative doesn't have gradient function

cut off the trace back : .detach()


```python
t = torch.ones(2,2,requires_grad = True)
print (t)
print (t.requires_grad)

y = t + 5
z = y * 2
out = z.mean()

print(out)
print (y.grad_fn)
print(z.grad_fn)
print (out.grad_fn)

#activate backward trace : backward can only be applied to a scalar (number, not valid to vector)
out.backward()

# after being activated, can track original step's gradient
print("d(out)/d(t):", t.grad)

result = out.detach()
print (result.requires_grad)


```
