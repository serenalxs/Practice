# **Initialize Tensor** : Int64, float32,float 64


1. create tensor directorly

```python
t = torch.tensor([1,2,3])
print(t)
print(t.dtype)

t = torch.FloatTensor([1,2,3])
print(t)
print(t.dtype)

t = torch.LongTensor([1,2,3])
print(t)
print(t.dtype)
```

2. create tensor from numpy

```python
arr = np.arange(12).reshape(3,4)
print(arr)

t = torch.from_numpy(arr)
print(t)

```
